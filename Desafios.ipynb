Desafio 06: Renomear as colunas tirando o hífen.

Utilizamos o metodo "rename" para renomear as colunas e passamos no paramentro "columns" uma funçao lambda que ira receber o valor da coluna e com o funçao "replace" retirar o hifen.
O parametro inplace como "True", do metodo "rename" ira aplicar as modificações.

```python
dados.rename(columns=lambda x: x.replace('-', ''), inplace=True)
dados
```

Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

```python
#carregamos o pacote
import matplotlib.pyplot as plt
```

```python
plt.pie(dados['tratamento'].value_counts(), labels =['Com Droga','Com Controle'], colors = ['#3A26F0','#0AA31E'], autopct='%1.2f%%')
plt.title("Tipo de Tratamento Submetido")
```

```python
dados['tempo'].value_counts().plot.bar(color = ['#0B403A','#178073','#22BFAD'], rot=0)
plt.title("Tempo de exposição a droga")
plt.xlabel("Tempo/hrs")
plt.ylabel("N.º de observações")
```

```python
dados['dose'].value_counts().plot.bar(color = ['#FF2401','#FFBA19'], rot=0)
plt.title("Quantidade de Doses")
plt.xlabel("Tipo de Dose")
plt.ylabel("Quantidade")
```

Desafio 01: Ordenar o gráfico countplot<br>
Desafio 02: Melhorar a visualização alterando tamanho da fonte..<br>
Desafio 05: Refletir sobre a manipulação do tamanho das visualizações.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
plt.figure(figsize=(8, 6))
cod_compostos = dados['composto'].value_counts().index[0:5]
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = cod_compostos)
ax.set_title('Top 5 compostos', fontsize=25)
plt.ylabel("Frequencia", fontsize=15)
plt.xlabel("Composto", fontsize=15)
plt.show()
```

Desafio 03: Plotar os histogramas com seaborn
